```python
import pandas as pd

cv = pd.read_csv("./sungbin/prbc_in_cv_210910.csv")
mv = pd.read_csv("./sungbin/prbc_in_mv_210910.csv")
```

## CV정리하기

```python
cv
```

```python
cv.columns
```

```python
drop_null_cv = cv[cv.AMOUNT.notnull()]
drop_null_cv
```

```python
drop_zero_null_cv = drop_null_cv[drop_null_cv.AMOUNT != 0.]
drop_zero_null_cv
```

```python
drop_zero_null_cv
```

# MV 정리하기

```python
mv
```

```python
mv.AMOUNT.value_counts()
```

```python
drop_zero_mv = mv[mv.AMOUNT != 0.]
drop_zero_mv
```

## 선별된 DRG_firstICU_DIAG와 inputevent MV, CV  머지하기

```python
drg_firstICU_DIAG = pd.read_csv('./sungbin/9_10_DRG_firstICU_DIAG_CV_MV.csv')
drg_firstICU_DIAG
```

```python
drop_dup_drg_firstICU_DIAG = drg_firstICU_DIAG[['SUBJECT_ID','HADM_ID','ICUSTAY_ID','label']].drop_duplicates()
drop_dup_drg_firstICU_DIAG
```

```python
drg_cv = pd.merge(drop_dup_drg_firstICU_DIAG, drop_zero_null_cv, on=['SUBJECT_ID','HADM_ID','ICUSTAY_ID'], how='inner')
drg_cv
```

```python
drg_mv = pd.merge(drop_dup_drg_firstICU_DIAG, drop_zero_mv, on=['SUBJECT_ID','HADM_ID','ICUSTAY_ID'], how='inner')
drg_mv
```

```python
cv_col = ['SUBJECT_ID','HADM_ID','ICUSTAY_ID','CHARTTIME','ITEMID','AMOUNT','AMOUNTUOM','label']
col_drg_cv= drg_cv[cv_col]
col_drg_cv = col_drg_cv.rename(columns={"CHARTTIME": "PRBC_TIME"})
col_drg_cv
```

```python
mv_col = ['SUBJECT_ID','HADM_ID','ICUSTAY_ID','STARTTIME','ITEMID','AMOUNT','AMOUNTUOM','label']
col_drg_mv = drg_mv[mv_col]
col_drg_mv = col_drg_mv.rename(columns={"STARTTIME": "PRBC_TIME"})
col_drg_mv
```

```python
final_cv_mv = pd.concat([col_drg_cv,col_drg_mv], axis=0)
final_cv_mv
```

```python
final_cv_mv = final_cv_mv.sort_values(by=['SUBJECT_ID', 'HADM_ID','ICUSTAY_ID'])
final_cv_mv
```

```python
final_cv_mv.to_csv("./sungbin/9_12_integrate_cv_mv.csv", index=False)
```
