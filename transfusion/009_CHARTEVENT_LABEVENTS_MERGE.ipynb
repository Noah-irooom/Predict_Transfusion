# MV CV 대상군 랩이벤트 차트이벤트 통합

```python
import pandas as pd
lab = pd.read_csv('./sungbin/9_19_LABEVENTS_ITEMID_20.csv')
chart = pd.read_csv('./sungbin/test0.csv')
```

```python
lab.rename(columns={'CAL_LAB_TIME' : 'CAL_TIME', 'MEAN_VALUENUM':'SCALED_VALUENUM'}, inplace = True)
# lab.drop(columns=['SUBJECT_ID','HADM_ID'], inplace = True)
lab
```

```python
chart.rename(columns={'CAL_CHART_TIME' : 'CAL_TIME'}, inplace = True)
chart
```

```python
new_merge = pd.merge(lab, chart, on=['SUBJECT_ID','HADM_ID','ICUSTAY_ID','CAL_TIME','ITEMID','SCALED_VALUENUM'], how='outer')
new_merge
```

```python
table = pd.pivot_table(new_merge, values='SCALED_VALUENUM', index=['SUBJECT_ID','HADM_ID','ICUSTAY_ID', 'CAL_TIME'],
                    columns=['ITEMID'])
table
```
